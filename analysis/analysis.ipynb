```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os 
```

```python
happo_df = pd.read_csv('./data/avg_rewards/anymal_c_multi_agent_happo.csv')
mappo_df = pd.read_csv('./data/avg_rewards/anymal_c_multi_agent_mappo.csv')
```

```python
num_steps = 100

happo_df_s = happo_df[:num_steps]
mappo_df_s = mappo_df[:num_steps]

plt.figure(figsize=(5,5))
sns.set_style('whitegrid')
sns.set_palette('colorblind')
sns.set_context('paper')
max_val_idx = np.argmax(happo_df_s['Value'].values)
max_val = happo_df_s['Value'].values[max_val_idx]
max_step_val = happo_df_s['Step'].iloc[max_val_idx]
sns.lineplot(happo_df_s, x='Step', y='Value', label=f'HAPPO')
plt.plot(max_step_val, max_val, color='r', marker='o', linestyle='None', label=f'HAPPO max reward {max_val:.2f} \nat step {max_step_val:,}')
max_val_idx = np.argmax(mappo_df_s['Value'].values)
max_val = mappo_df_s['Value'].values[max_val_idx]
max_step_val = mappo_df_s['Step'].iloc[max_val_idx]
sns.lineplot(mappo_df_s, x='Step', y='Value', label=f'MAPPO')
plt.plot(max_step_val, max_val, color='r', marker='x', linestyle='None', label=f'MAPPO max reward {max_val:.2f} \nat step {max_step_val:,}')
sns.despine()
plt.ylabel('Reward')
plt.legend()
plt.show()
```
