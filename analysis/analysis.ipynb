```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os 
```

```python
happo_df = pd.read_csv('./data/avg_rewards/anymal_c_multi_agent_happo.csv')
mappo_df = pd.read_csv('./data/avg_rewards/train_episode_rewards_aver_rewards(1).csv')
```

```python
sns.set_style('whitegrid')
sns.set_palette('colorblind')
sns.set_context('paper')
sns.lineplot(happo_df[:30], x='Step', y='Value', label='HAPPO')
sns.lineplot(mappo_df, x='Step', y='Value', label='MAPPO')
sns.despine()
plt.ylabel('Reward')
plt.legend()
plt.show()
```
